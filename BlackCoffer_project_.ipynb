```python
def extract_text(df,output_folder):

  if not os.path.exists(output_folder):
    os.makedirs(output_folder)

  processed_rows = []  # To keep track of processed rows

  for i in range(df.shape[0]):
    url_id = df.iloc[i,0]
    url = df.iloc[i,1]
    response = rq.get(url)
    file_name = os.path.join(output_folder, f"{url_id}.txt")

    if response.status_code == 200:
      soup_data = bs(response.text, 'html.parser')
      processed_rows.append(url_id)



      #for div in (["tdb-block-inner td-fix-index", "td-post-content tagdiv-type"]):
      div_element = soup_data.select_one('div.td-post-content.tagdiv-type')
      if div_element is not None:
          # Find all <p> and <ol> elements within div_element in order of appearance
        mixed_elements = div_element.find_all(lambda tag: tag.name in ['p', 'ol','ul'])


        article_text = ""
        for p in mixed_elements:
          article_text += p.get_text() + "\n"

        with open(file_name, "w") as text_file:
          text_file.write(article_text)

  # Print the rows for which files were not created
  not_processed_rows = [url_id for url_id in df['URL_ID'] if url_id not in processed_rows]
  if not_processed_rows:
      print(f"Files were not created for the following rows: {not_processed_rows}")
      print(f"No. files created:{len(processed_rows)}")
      print(f"Total Rows for files:{df.shape[0]}")
  else:
    print("All rows processed successfully.")
```

```python
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import syllables
```

```python
def calculate_metrices(text):
  blob = TextBlob(text)      #undestanding
  word_count = len(blob.words)    #total no. of words in text
  sentence_count = len(blob.sentences)    #total no. of sentences in text
  avg_sentence_length = (word_count/sentence_count)if sentence_count != 0 else 0
  complex_word_count = len([word for word in blob.words if syllables.estimate(word)>=3])  #complex_word = (no. of syllables>=3)
  percentage_complex_words = ((complex_word_count/word_count)*100) if word_count != 0 else 0
  fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

  syllable_per_word = (sum(syllables.estimate(word) for word in blob.words) / word_count)if word_count != 0 else 0
  personal_pronoun = sum(1 for word, pos in blob.tags if pos == 'PRP')
  avg_word_length = (sum(len(word) for word in blob.words) / word_count)if word_count != 0 else 0

  # Initialize the VADER sentiment analyzer
  sia = SentimentIntensityAnalyzer()
  # Get the sentiment scores for the paragraph
  sentiment_scores = sia.polarity_scores(text)

  # Extract positive and negative scores
  positive_score = sentiment_scores['pos']
  negative_score = sentiment_scores['neg']

  return {
        'POSITIVE SCORE': round(positive_score,2),
        'NEGATIVE SCORE': round(negative_score,2),
        'POLARITY SCORE': round(blob.sentiment.polarity,2),
        'SUBJECTIVITY SCORE': round(blob.sentiment.subjectivity,2),
        'AVG SENTENCE LENGTH': round(avg_sentence_length,2),
        'PERCENTAGE OF COMPLEX WORDS': round(percentage_complex_words,2),
        'FOG INDEX': round(fog_index,2),
        'AVG NUMBER OF WORDS PER SENTENCE': round(avg_sentence_length,2),
        'COMPLEX WORD COUNT': round(complex_word_count,2),
        'WORD COUNT': round(word_count,2),
        'SYLLABLE PER WORD': round(syllable_per_word,2),
        'PERSONAL PRONOUNS': round(personal_pronoun,2),
        'AVG WORD LENGTH': round(avg_word_length,2)
    }
```

```python
import pandas as pd
import os
```

```python
# Path to the folder containing TXT files in your Google Drive
folder_path = '/content/drive/My Drive/black_coffer/'
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
# DataFrame to store the results
results_df = pd.DataFrame()

# Iterate through files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    with open(file_path, 'r') as file:
        data = file.read()

    # Calculate metrics for the text in the file
    metrics = calculate_metrices(data)

    # Append the results to the DataFrame
    results_df = results_df.append({'URL_ID': filename, **metrics}, ignore_index=True)
```

```python
results_df_=results_df.sort_values(by=['URL_ID'])
results_df_.head(2)
```
